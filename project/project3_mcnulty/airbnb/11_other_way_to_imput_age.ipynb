```python
import pickle

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
with open('./data/processed/X_user_train.pkl', 'rb') as picklefile:
    user = pickle.load(picklefile)
```

```python
user = user[(user['age'] > 10) & (user['age'] < 100) ]
```

```python
user.shape
```

```python
user.first_affiliate_tracked.fillna('catch-all', inplace=True)
```

```python
COLS_TO_VISUALIZE = ['gender', 'signup_method', 'signup_flow', 'affiliate_channel','first_affiliate_tracked',  'signup_app', 'first_device_type', 'first_browser', 'account_creation_year', 'active_year']
```

```python
user.dropna(axis =0, subset = ['age'], inplace=True)
```

```python
user['account_creation_year'] = user['date_account_created'].apply(lambda x:x.year)
user['active_year'] = user['timestamp_first_active'].apply(lambda x:x.year)
```

```python
for col in COLS_TO_VISUALIZE:
    plt.figure()
    ax = user.boxplot(column='age', by=col, rot=90)
    plt.xlabel(col)
    plt.ylabel('age')
    plt.title(col)
```

```python
plt.figure()
ax = user.boxplot(column='age', by=['affiliate_channel', 'gender'], rot=90)
plt.xlabel(col)
plt.ylabel('age')
plt.title(col)
```

```python
plt.figure()
ax = user.boxplot(column='age', by=['first_device_type', 'gender'], rot=90)
plt.xlabel(col)
plt.ylabel('age')
plt.title(col)
```
