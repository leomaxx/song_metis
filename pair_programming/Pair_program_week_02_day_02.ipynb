```python
import numpy as np

a = np.array([2,7,1,5,10])
b = np.arange(0,10,0.01)

results= np.zeros((1000,1))

for i in range(b.shape[0]):
    sum =0
    for y in a:
        sum = sum + (b[i]-y)**2
    results[i] = sum
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(b, results)
plt.plot()
```

```python
b[results.argmin()]
```

```python
results.min()
```

```python
for i in range(b.shape[0]):
    diff = b[i] - a
    results[i] = np.sum(np.multiply(diff, diff))
    
```

```python
results.min()
```

```python
plt.plot(b, results)
plt.plot()
```
